```python
import pandas as pd
from sqlalchemy import create_engine
```

## Store CSV into DataFrame

```python
hospitals_df = "Resources/Minnesota_Hospitals.csv"
hospitals_df = pd.read_csv(hospitals_df)
hospitals_df.head()
```

## For reading this file we had to use a special code for decoding "ISO-8859-1"

```python
mndivs_df = "Resources/indivs_Minnesota18 2.csv"
mndivs_df = pd.read_csv(mndivs_df, encoding="ISO-8859-1")
mndivs_df.head()
```

## Transform the hospital_df file

```python
# Transform the hospital_df file
hospitals_cols_df = ["OBJECTID", "NAME", "ADDRESS","CITY","STATE","HOSP_BEDS"]
hospitals_transformed= hospitals_df[hospitals_cols_df].copy()

# Rename the column headers
hospitals_transformed = hospitals_transformed.rename(columns={"OBJECTID": "HOSPITAL_ID",
                                                          "NAME": "NAME",
                                                          "ADDRESS": "ADDRESS",
                                                          "CITY": "CITY",
                                                          "STATE": "STATE",
                                                          "HOSP_BEDS": "HOSP_BEDS"
                                                             })

# Clean the data by dropping duplicates and setting the index
hospitals_transformed.drop_duplicates("HOSPITAL_ID", inplace=True)
hospitals_transformed.set_index("HOSPITAL_ID", inplace=True)

hospitals_transformed.head()
```

## Transform the mndivs_df file

```python
mndivs_df.columns
```

```python
# Create a filtered dataframe from specific columns
mndivs_cols = ["Contribid", "Orgname", "Amount","City","State","Zip","Gender"]
mndivs_transformed= mndivs_df[mndivs_cols].copy()

# Rename the column headers
mndivs_transformed = mndivs_transformed.rename(columns={"Contribid": "Contribution_ID",
                                                          })

# Clean the data by dropping duplicates and setting the index
mndivs_transformed.drop_duplicates("Contribution_ID", inplace=True)
mndivs_transformed.set_index("Contribution_ID", inplace=True)

mndivs_transformed
```

## Create database connection


```python
connection_string = "postgres:bolboceanu13na@localhost:5432/etl_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
# Confirm tables
engine.table_names()
```

## Load DataFrames into database

```python
hospitals_transformed.to_sql(name='mnhospital', con=engine, if_exists='append', index=True)
```

```python
mndivs_transformed.to_sql(name='mndivs3', con=engine, if_exists='append', index=False)
```
